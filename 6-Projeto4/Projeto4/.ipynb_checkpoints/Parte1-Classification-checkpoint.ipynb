### Definção do Problema

O câncer de pulmão é um dos tipos mais comuns de câncer e foi responsável por mais de 1,5 milhão de mortes em todo o mundo em 2012, segundo a Organização Mundial de Saúde. Usando tomografias computadorizadas, é possível avaliar se as pessoas correm risco de potencialmente desenvolver câncer de pulmão ou outras doenças pulmonares usando Machine Learning. No entanto, a tecnologia atual ainda é limitada e produz muitos falsos positivos que levam a exames de acompanhamento desnecessários e ansiedade do paciente. Recentemente, resultados promissores foram obtidos para problemas semelhantes como detecção de câncer de pele. 

A fim de melhorar o estado atual da detecção de câncer de pulmão, o Kaggle lançou uma competição no valor de $1.000.000 em prêmios totais para o Data Science Bowl 2017, onde o objetivo era desenvolver algoritmos para determinar se um paciente desenvolverá câncer ou não. A competição tinha como objetivo reunir a comunidade científica e médica para dar grandes saltos nos algoritmos de detecção de câncer, mais especificamente detecção precoce. Detecção precoce do câncer de pulmão é importante porque permite um tratamento antecipado e tem potencial para reduzir as mortes.

A forma como o câncer de pulmão é diagnosticado é inspecionando as imagens de tomografia computadorizada de um paciente, procurando pequenas bolhas nos pulmões chamadas nódulos. Encontrar um nódulo em si não é um indicativo de câncer; os nódulos têm que ter certas características (forma, tamanho, etc.) para apoiar um diagnóstico de câncer.

### Soluções Propostas Por Pesquisadores

Várias abordagens para a solução da detecção e classificação de nódulos pulmonares foram propostas na literatura, incluindo as duas técnicas que dependem de ferramentas tradicionais de processamento de imagens e de visão computacional, e aquelas que dependem de aprendizagem profunda (Deep Learning). A detecção de nódulos geralmente depende de técnicas de segmentação de imagem. Técnicas tradicionais de processamento de imagens fazem uso de características de intensidade e forma geométrica de componentes conectados, cortes de grafos e contornos ativos, e clusters c-means. Recentemente, as redes convolucionais de arquitetura U-Net mostraram-se boas para segmentar estruturas e identificar regiões com nódulos. 

Uma característica distintiva dessa arquitetura de rede é que combina recursos de baixo nível nas camadas iniciais com  recursos upsampled de nível superior nas camadas finais para fazer suas previsões. A abordagem é muito mais rápida do que aquelas em que um classista é treinado para prever cada pixel baseado no patch circundante, uma vez que não requer que a rede faça previsões no nível de pixel. Outras abordagens dependem de informações 3D para detecção, como usar uma CNN 3D para classificar subvolumes como contendo um nódulo ou não, ou usando uma abordagem de classificação no nível do voxel com uma CNN 3D. Para a classificação dos nódulos segmentados, os métodos relatados utilizaram abordagem baseada em autoencoder com uma árvore de decisão binárias, características profundas de uma CNN passadas para os classificadores SVM e Random Forest, o Adaboost classier, enquanto as abordagens anteriores dependiam da computação manual características de forma, textura e tamanho de nódulos segmentados com classificadores superficiais como LDA, discriminante quadrático e regressão logística.

### Nossa Solução

A solução que apresenta o melhor resultado é a combinação de técnicas de Visão Computacional e Deep Learning, usando uma arquitetura U-Net, solução que usaremos neste projeto.

Começamos por detectar nódulos de imagens CT de cada paciente e extrair os patches ao redor deles. Então nós codificamos esses patches usando uma das três codificações (a serem descritas a seguir) e usamos a codificação média de todos os patches retornados para um paciente como o vetor de recurso representando-os. Por fim, treinamos um classificador SVM nesses vetores de recursos.

A detecção do nódulo é feita usando a U-Net. Para reduzir o número de falsos positivos retornados pela U-Net, seguimos um heurística simples nesta abordagem: para qualquer nódulo em qualquer fatia, se a U-Net não detectar um nódulo próximo a fatia (com o conjunto de fatias ordenadas por posição) na vizinhança daquele nódulo, descartamos. A suposição aqui é que os nódulos se estendem mais de várias fatias e, consequentemente, um nódulo que aparece em apenas uma fatia é assumido como um falso positivo. Extraímos patches de 48 x 48 em torno dos nódulos candidatos. Em seguida, usamos Autoencoders, local binary patterns e flattening para codificar esses patches.

### Autoencoders

Autoencoders são um tipo especial de redes neurais que são treinadas para mapear a entrada em si, com a restrição de que uma das camadas ocultas (geralmente chamado de camada bottleneck) tem uma menor dimensionalidade do que a entrada. Nesse sentido, eles representam uma classe de técnicas de redução de dimensionalidade. 

Um convoutional autoencoder é um autoencoder que utiliza camadas convolucionais (usando o início do codificador e o final do decodificador). Nós usamos um convolucional autoencoder e treinamos em todos os patches extraídos de todos os pacientes. Então, Usamos o autoencoder para codificar cada patch 48 x 48 como um vetor 64-dimensional. ### Local Binary Patterns

Padrões Binários Locais é um descritor popular em Visão Computacional usado para capturar o conteúdo textural das imagens. Cada pixel é um código binário baseado em seu valor em relação aos seus pixels vizinhos. Isso é feito atribuindo um valor de 1 aos pixels que possuem vizinhos com valores maiores ou iguais ao valor do pixel sendo considerado e um valor de zero para aqueles com valores mais baixos. Isto resulta em um código binário de n bits que descreve cada pixel, onde n é o número de vizinhos. Este processo é ilustrado na figura abaixo. O histograma desses códigos binários em uma imagem é usada como o vetor de recursos que descreve essa imagem.

### Flattening

A última abordagem para codificar os patches candidatos é usar o pixels de patch como a codificação. Em outras palavras, nós meramente aplicamos o flatten em 48 x 48 patches em vetores. 

Esses três métodos fornecem codificações para os patches. Dos três métodos, usando o representação vetorial flattening produziu os melhores resultados em conjunto de dados. Contudo, no final, precisamos de uma representação no nível do paciente; ou seja, precisamos de um único vetor de características descrevendo cada paciente, já que os rótulos descritos descrevem paciente é diagnosticado com câncer ou não, mas não diz nada sobre nódulos. Para este fim, usamos a codificação média de todos os patches detectados para um dado um paciente como o vetor de característica final representando esse paciente. Finalmente, nós treinamos um classificador de máquina de vetores de suporte (SVM) com função de base radial (RBF) kernel nesses vetores de recursos.

### Abordagem Final

Nossa abordagem final envolve manter os nódulos detectados que são encontrados em regiões onde foram feitas várias detecções em todas as fatias para um único paciente. Essa abordagem gera um mapa de calor para combinar as detecções da U-Net em todas as fatias de um único paciente.

A figura abaixo mostra todas as regiões dos nódulos detectados para um paciente com as regiões que são mais vermelhas, indicando mais detecções nesses locais de pixel. Pode ser visto que existem detecções de falso positivo em vários locais da tomografia computadorizada, mas apenas algumas regiões têm múltiplas detecções, ou seja, as bolhas vermelhas com maior intensidade. O mapa de calor é então limitado para manter apenas os pixels que possuem repetidas detecções. 

A ideia por trás dessa abordagem é que as regiões que tiveram várias detecções da segmentação do nódulo são mais propensos a ser nódulos reais. A partir daí, apenas as fatias que se sobrepõem ao limiar das regiões do nódulo do mapa de calor são mantidas. ## Preparação dos Dados

Começamos detectando as regiões candidatas de nódulo. A fim de reduzir o número de falsos positivos gerados neste processo, é útil isolar as regiões do pulmão (onde aparecem os nódulos) e limpar todo o resto nas tomografias. Fazemos isso seguindo o tutorial do Kaggle mencionado anteriormente. A abordagem descrita no tutorial se baseia em heurísticas que parecem funcionar bem nos conjuntos de dados Kaggle e LUNA. A etapa de segmentação pulmonar isola os pulmões por limiar as imagens com base na intensidade dos pixels. Em seguida, a filtragem morfológica (dilatação e erosão) é usada para eliminar a máscara pulmonar, remendando quaisquer orifícios que possam resultar do thresholding.

A normalização do escore Z é usada. Cada imagem na tomografia computadorizada é centrada pela subtração de sua média e normalizada pela divisão por sua desvio padrão. Nós achamos que essa abordagem funciona melhor do que normalizar todo o conjunto de dados como um todo (usando a média e o desvio padrão do conjunto de dados inteiro).

```python
# Preparando os dados para treinamento
nodules = data[0][0]
labels = np.ones(data[0][0].shape[0]) * data[0][2]
for idx, patient in enumerate(data[1:]):
    if patient[0].any():
        labels = np.concatenate((labels, np.ones(patient[0].shape[0]) * patient[2]))
        nodules = np.vstack((nodules, patient[0]))
        
num_classes = 2
num_samples = nodules.shape[0]
img_rows = nodules.shape[1]
img_cols = nodules.shape[2]

X_train, X_test, y_train, y_test = train_test_split(nodules, labels, test_size=0.33, random_state=42)

X_train = X_train.reshape(X_train.shape[0], img_rows, img_cols, 1)
X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols, 1)
input_shape = (img_rows, img_cols, 1)

y_train = np_utils.to_categorical(y_train, num_classes)
y_test = np_utils.to_categorical(y_test, num_classes)         
```
