```python
import pandas
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error 
```

```python
dataset = pandas.read_csv('winemag-data_first150k.csv')
```

```python
dataset.head(5)
```

```python
sw = ["to","of","it","I","the","is", "up", "has", "for", "have", "yet", "I've", "a", "as", "this"]
vectorizer = TfidfVectorizer(stop_words = sw)
```

```python
X = vectorizer.fit_transform(dataset.description)
```

```python
vectorizer.vocabulary_
```

### Treinando o modelo

```python
X_train, X_test, y_train, y_test = train_test_split(X, dataset.points, test_size=0.2, random_state=42)
```

### Usando Linear Regression 

```python
reg = linear_model.LinearRegression()
reg.fit(X_train, y_train)
```

## Fazendo a predição

```python
y_pred = reg.predict(X_test)
y_pred
```

## Avaliando o modelo

```python
mean_squared_error(y_test, y_pred)
```
